```python
# Function to transform scan using pose
def transform_doppler_points(transform, dpoints):
    # N x [x y z rcs v v_comp]
    result_dpoints = np.zeros_like(dpoints)

    points_h = np.ones((dpoints.shape[0], 4))
    points_h[:, :3] = dpoints[:, :3]
    points_h = (points_h @ transform.T)
    result_dpoints[:, :3] = points_h[:, :3]

    if dpoints.shape[1] > 3:
        result_dpoints[:, 3:] = dpoints[:, 3:]

    return result_dpoints

# Color map for labels
color_map = {
    0 : np.array([0, 1, 0]).astype(float),  # Green for class 0
    1 : np.array([1, 0, 0]).astype(float),  # Red for class 1
}

# Function to plot each frame
def plot_frame(i):
    fig, axs = plt.subplots(1, 3, figsize=(24, 8))
    
    # GT Plot (First Column)
    scan = np.load(os.path.join(scans_dir, filenames[i]))
    labels = np.load(os.path.join(labels_dir, filenames[i]))
    labels = (labels <= 0.5).astype(int) # NOTE Plotting for unstable
    pose = np.loadtxt(os.path.join(poses_dir, filenames[i].replace('.npy', '.txt')), delimiter=',')
    transformed_scan = transform_doppler_points(pose, scan)
    colors = [color_map[l] for l in labels]
    axs[0].scatter(transformed_scan[:, 0], transformed_scan[:, 1], c=colors, s=1)
    axs[0].set_title(f"Ground Truth ({gt_dir.split('/')[-2]}): {filenames[i]}")
    axs[0].set_xlabel('X')
    axs[0].set_ylabel('Y')
    
    # Experiment 1 (Second Column)
    preds_exp1 = np.load(os.path.join(predictions_dir_exp1, pred_filenames_exp1[i]))
    coords_exp1 = preds_exp1[:,:3]
    sps_scores_exp1 = preds_exp1[:,-1]
    labels_exp1 = [int(s > sps_thresh) for s in sps_scores_exp1]
    colors_exp1 = [color_map[l] for l in labels_exp1]
    axs[1].scatter(coords_exp1[:, 0], coords_exp1[:, 1], c=colors_exp1, s=1)
    axs[1].set_title(f"SPS: {pred_filenames_exp1[i]}")
    axs[1].set_xlabel('X')
    axs[1].set_ylabel('Y')

    # Experiment 2 (Third Column)
    preds_exp2 = np.load(os.path.join(predictions_dir_exp2, pred_filenames_exp2[i]))
    coords_exp2 = preds_exp2[:,:3]
    sps_scores_exp2 = preds_exp2[:,-1] # Stable to unstable if needed
    # labels_exp2 = [int(s > rit_thresh) for s in sps_scores_exp2]
    
    # labels_exp2 = 1 - preds_exp2[:,-1] # Stable to unstable if needed (output from test.py)
    labels_exp2 = preds_exp2[:,-3]

    colors_exp2 = [color_map[l] for l in labels_exp2]
    axs[2].scatter(coords_exp2[:, 0], coords_exp2[:, 1], c=colors_exp2, s=1)
    axs[2].set_title(f"RIT: {pred_filenames_exp2[i]}")
    axs[2].set_xlabel('X')
    axs[2].set_ylabel('Y')

    plt.tight_layout()
    plt.show()

# Create a slider widget to view different frames
slider = IntSlider(min=0, max=len(filenames)-1, step=1, description='Frame')
interact(plot_frame, i=slider)

```

```python
import numpy as np

def update_pointcloud_pose(pointcloud, from_pose, to_pose):
    # Calculate the transformation matrix (inverse of to_pose multiplied by from_pose)
    transformation = to_pose @ np.linalg.inv(from_pose)
    
    # Number of points in the point cloud (N x 4)
    NP = pointcloud.shape[0]
    
    # Separate the point cloud into xyz (N x 3) and the labels (N x 1)
    xyz = pointcloud[:, :3]
    labels = pointcloud[:, 3].reshape(NP, 1)  # Keep the labels intact
    
    # Add a column of ones to the xyz coordinates for homogeneous transformation (N x 4)
    xyz1 = np.hstack([xyz, np.ones((NP, 1))]).T  # Shape: (4, N)
    
    # Apply the transformation
    transformed_xyz = (transformation @ xyz1).T[:, :3]  # Shape: (N, 3)
    
    # Concatenate the transformed xyz with the labels (N x 4)
    transformed_point_clouds = np.hstack([transformed_xyz, labels])
    
    return transformed_point_clouds


def plot_predicted_maps(gt_pcl, sps_pcl, rit_pcl, scene_name):
    import matplotlib.pyplot as plt
    # Create subplots with 1 row and 2 columns
    fig, axes = plt.subplots(1, 3, figsize=(24, 8))
    
    gt_labels = [int(s > 0.5) for s in gt_pcl[:, -1]]
    axes[0].scatter(gt_pcl[:, 0], gt_pcl[:, 1], c=gt_labels, cmap='RdYlGn', s=0.1)
    axes[0].set_title(f'Ground-truth {scene_name}')
    axes[0].set_xlabel('X')
    axes[0].set_ylabel('Y')

    # First point cloud in the left subplot
    sps_labels = [int(s > 0.84) for s in sps_pcl[:, -1]]
    axes[1].scatter(sps_pcl[:, 0], sps_pcl[:, 1], c=sps_labels, cmap='RdYlGn', s=0.1)
    axes[1].set_title(f'SPS predictions {scene_name}')
    axes[1].set_xlabel('X')
    axes[1].set_ylabel('Y')
    
    # Second point cloud in the right subplot
    rit_labels = [int(s) for s in rit_pcl[:, -1]]
    axes[2].scatter(rit_pcl[:, 0], rit_pcl[:, 1], c=rit_labels, cmap='RdYlGn', s=0.1)
    axes[2].set_title(f'RIT predictions {scene_name}')
    axes[2].set_xlabel('X')
    axes[2].set_ylabel('Y')
    
    # Adjust layout
    plt.tight_layout()
    plt.show()


def parse_rit_predictions(dataset_dir, predictions_dir, scene_name):
    src_pose_dir = os.path.join(dataset_dir, scene_name, 'local_poses')
    dst_pose_dir = os.path.join(dataset_dir, scene_name, 'poses')
    scans_dir = os.path.join(dataset_dir, scene_name, 'scans')

    scan_files = sorted([os.path.join(scans_dir, file) for file in os.listdir(scans_dir)] , key=lambda f: float(''.join(filter(lambda x: x.isdigit() or x == '.', f.split('.npy')[0]))))
    src_pose_paths = sorted([os.path.join(src_pose_dir, file) for file in os.listdir(src_pose_dir)] , key=lambda f: float(''.join(filter(lambda x: x.isdigit() or x == '.', f.split('.txt')[0]))))
    dst_pose_paths = sorted([os.path.join(dst_pose_dir, file) for file in os.listdir(dst_pose_dir)] , key=lambda f: float(''.join(filter(lambda x: x.isdigit() or x == '.', f.split('.txt')[0]))))
    
    src_poses = [np.loadtxt(pth, delimiter=',') for pth in src_pose_paths]
    dst_poses = [np.loadtxt(pth, delimiter=',') for pth in dst_pose_paths]

    all_scans = []
    for i,scan_file in enumerate(scan_files):
        predicted_pcl = np.load(os.path.join(predictions_dir, scene_name, f'{i}.npy'))

        global_predicted_pcl = update_pointcloud_pose(predicted_pcl, src_poses[i], dst_poses[i])
        all_scans.append(global_predicted_pcl)
        # output_map.append(predicted_pcl)
    
    output_map = np.vstack(all_scans)
    return output_map, all_scans


def parse_sps_predictions(dataset_dir, predictions_dir, scene_name):
    scans_dir = os.path.join(dataset_dir, scene_name, 'scans')
    scan_files = sorted([os.path.join(scans_dir, file) for file in os.listdir(scans_dir)] , key=lambda f: float(''.join(filter(lambda x: x.isdigit() or x == '.', f.split('.npy')[0]))))
    all_scans = []
    for i,scan_file in enumerate(scan_files):
        predicted_pcl = np.load(os.path.join(predictions_dir, scene_name, 'scans', f'{i}.npy'))[:, [0,1,2,4]]
        all_scans.append(predicted_pcl)
    
    output_map = np.vstack(all_scans)
    return output_map, all_scans
```

```python
## Used by SPS
def calculate_metrics(true_labels, predicted_labels):
    # Calculate True Positives, False Positives, False Negatives
    true_positives = np.sum(np.logical_and(true_labels == 1, predicted_labels == 1))
    true_negatives = np.sum(np.logical_and(true_labels == 0, predicted_labels == 0))
    false_positives = np.sum(np.logical_and(true_labels == 0, predicted_labels == 1))
    false_negatives = np.sum(np.logical_and(true_labels == 1, predicted_labels == 0))

    # Calculate Precision, Recall, and F1
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) != 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) != 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0
    accuracy = (true_positives + true_negatives) / (true_positives + true_negatives + false_positives + false_negatives)
    dIoU = true_positives / (true_positives + false_negatives + false_positives)

    return precision, recall, f1, accuracy, dIoU
    

## Used by RIT
def compute_iou_pointcloud(output, target, num_classes, ignore_index=255):
    """
    Computes class-wise IoU for point cloud segmentation.

    Args:
    - output (numpy array): Predicted labels for the point cloud (N,).
    - target (numpy array): Ground truth labels for the point cloud (N,).
    - num_classes (int): The number of classes (e.g., 2 for binary segmentation).
    - ignore_index (int): Label to ignore in both output and target.

    Returns:
    - iou_per_class (numpy array): Class-wise IoU for each class.
    """
    # Ensure output and target have the same shape
    assert output.shape == target.shape

    # Flatten output and target arrays if needed
    output = output.reshape(-1).copy()  # Make a copy to avoid modifying original data
    target = target.reshape(-1)

    # Mask out the ignored indices in the output
    output[target == ignore_index] = ignore_index

    # Compute intersection: where output equals target
    intersection = output[output == target]

    # Histogram binning to calculate the number of occurrences for each class
    area_intersection, _ = np.histogram(intersection, bins=np.arange(num_classes + 1))
    area_output, _ = np.histogram(output, bins=np.arange(num_classes + 1))
    area_target, _ = np.histogram(target, bins=np.arange(num_classes + 1))

    # Union is the sum of the target and predicted areas minus the intersection
    area_union = area_output + area_target - area_intersection

    # Compute IoU for each class: intersection over union
    iou_per_class = np.zeros(num_classes)
    for cls in range(num_classes):
        if area_union[cls] > 0:  # Avoid division by zero
            iou_per_class[cls] = area_intersection[cls] / area_union[cls]

    return iou_per_class
```

## Inspecting Tensorboard Logs

```python
import os
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

class TensorBoardReader:
    def __init__(self, log_dir):
        """
        Initialize the TensorBoardReader with the directory of the log file.
        Args:
        - log_dir (str): Path to the TensorBoard event log directory or file.
        """
        self.log_dir = log_dir
        self.data = {}

    def load_data(self):
        """
        Loads the data from the TensorBoard event log file.
        """
        # Initialize EventAccumulator to read the log file
        event_accumulator = EventAccumulator(self.log_dir)
        event_accumulator.Reload()  # Load the event file

        # Extract scalars for each metric and store in the dictionary
        tags = event_accumulator.Tags().get('scalars', [])
        
        for tag in tags:
            # Extract scalar events for the tag (metric name)
            events = event_accumulator.Scalars(tag)
            
            # Store the scalar values and their steps in the dictionary
            self.data[tag] = [(event.step, event.value) for event in events]

    def get_data(self):
        """
        Returns the extracted data as a dictionary.
        """
        if not self.data:
            self.load_data()
        return self.data
    
    
    def plot_metrics(self, keys, x_label='Epoch', y_label='Value'):
        """
        Plots the data corresponding to the list of keys in subplots.

        Args:
        - keys (list of str): List of metric names (keys) to plot.
        - x_label (str): The label for the x-axis (either 'Step' or 'Epoch').
        - y_label (str): The label for the y-axis (Value).
        """
        if not self.data:
            self.load_data()

        # Get the total number of keys and determine the subplot grid
        num_keys = len(keys)
        num_cols = 3  # Maximum of 3 columns per row
        num_rows = (num_keys + num_cols - 1) // num_cols  # Calculate the number of rows

        # Create the subplots dynamically based on the number of keys
        fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
        axes = axes.flatten()  # Flatten in case of multiple rows/columns

        for i, key in enumerate(keys):
            if key in self.data:
                steps, values = zip(*self.data[key])  # Unpack the steps and values
                axes[i].plot(steps, values, label=key)

                # Set title, labels, and grid
                axes[i].set_title(f'{key}', fontsize=14)
                axes[i].set_xlabel(x_label, fontsize=12)
                axes[i].set_ylabel(y_label, fontsize=12)
                axes[i].grid(True)
                axes[i].legend()

        # Remove unused subplots if there are fewer keys than axes
        for j in range(i + 1, len(axes)):
            fig.delaxes(axes[j])

        plt.tight_layout()
        plt.show()
```
